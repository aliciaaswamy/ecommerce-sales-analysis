```python
import pandas as pd
import mysql.connector

def load_csv_to_mysql(table_name, file_path):
    # Read CSV
    df = pd.read_csv(file_path)
    
    # Create fresh connection
    conn = mysql.connector.connect(
        host='localhost',
        user='root',
        password='root123',
        database='ecommerce'
    )
    cursor = conn.cursor()
    
    # Drop table if exists
    cursor.execute(f"DROP TABLE IF EXISTS {table_name}")
    
    # Create columns dynamically
    cols = []
    for col in df.columns:
        cols.append(f"`{col}` TEXT")
    create_stmt = f"CREATE TABLE {table_name} ({', '.join(cols)})"
    cursor.execute(create_stmt)
    
    # Insert rows
    for _, row in df.iterrows():
        values = tuple(str(v) if pd.notna(v) else None for v in row)
        placeholders = ', '.join(['%s'] * len(row))
        cursor.execute(f"INSERT INTO {table_name} VALUES ({placeholders})", values)
    
    conn.commit()
    cursor.close()
    conn.close()
    print(f" {table_name} loaded — {len(df)} rows")

# Load all tables
load_csv_to_mysql('customers', 'C:/ecommerce_project/data/olist_customers_dataset.csv')
load_csv_to_mysql('orders', 'C:/ecommerce_project/data/olist_orders_dataset.csv')
load_csv_to_mysql('order_items', 'C:/ecommerce_project/data/olist_order_items_dataset.csv')
load_csv_to_mysql('order_payments', 'C:/ecommerce_project/data/olist_order_payments_dataset.csv')
load_csv_to_mysql('order_reviews', 'C:/ecommerce_project/data/olist_order_reviews_dataset.csv')
load_csv_to_mysql('products', 'C:/ecommerce_project/data/olist_products_dataset.csv')
load_csv_to_mysql('sellers', 'C:/ecommerce_project/data/olist_sellers_dataset.csv')
load_csv_to_mysql('geolocation', 'C:/ecommerce_project/data/olist_geolocation_dataset.csv')
load_csv_to_mysql('category_translation', 'C:/ecommerce_project/data/product_category_name_translation.csv')

print("\n All tables loaded successfully!")
```
